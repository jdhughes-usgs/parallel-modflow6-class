## Parallel settings

1. Set simulation_type to a valid simulation type (listed above).
2. Set metis to `True` to evaluate performance of structured grid base and parallel models split with Metis. metis can be `True` or `False` is processing an unstructured grid.
3. set ksp_type to `cg` for Conjugate Gradient and `bcgs` for Bi-Conjugate Gradient Stabilized.
4. set dvclose to a real value to define the closure criteria for the dependent variable in PETSc. dvclose should be smaller that the outer_dvclose set in the IMS solver.
5. Set use_gamg to `True` to use the Geometric Algebaic Multi-Grid parallel preconditioner. If use_gamg is `False`, the Block Jacobi parallel preconditioner will be used.

```python
simulation_type = "basin_structured"
metis = True
ksp_type = 'bcgs'
dvclose = 1e-06
use_gamg = False
```

```python
paths = get_available_workspaces(metis=metis, simulation_type=simulation_type)
paths
```

```python
for path in paths:
    write_petscdb(path, ksp_type=ksp_type, dvclose=dvclose, use_gamg=use_gamg,)
```
